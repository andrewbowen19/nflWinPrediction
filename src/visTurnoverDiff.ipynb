### Notebook to visualize yearly turnover differential change for each NFL team

```python
import pandas as pd
import plotly.express as px
import numpy
import os
import matplotlib.pyplot as plt
```

```python
# Reading in and formatting df
df = pd.read_csv("turnover_diff_by_team_by_season.csv").T
df.columns = df.iloc[0]
df = df.iloc[1::]
```

```python
# Data goes back to 2003
df.head()
```

```python
# Interactive plotly line graph -- a bit busy with every team plotted
# COuld be fun to write a web app with which teams suck the most
px.line(df, x=df.index, y=df.columns)
```

```python
print(df.index[::-1])
```

```python
# Plotting each team's season

for c in df.columns:
    print(c)
    f,ax = plt.subplots()
    ax.plot(df.index[::-1], df[c])
    ax.set_title(c)
#     ax.set_xticklabels(ax.get_xticks(), rotation = 45)
#     print(ax.get_xticklabels())
```
